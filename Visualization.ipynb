```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv("C:\MyCodes\PYTHON\Data Analysis\cleaned_data_BFP_FIreIncidents2012-2016.csv")
```

# Line Chart

```python
fig1 = px.line(df, x="YEAR", y="INCIDENTS",
               title="Incidents Over the year",
               markers=True,
               color = 'REGION')
fig1.show()
```

# Bar chart of total incidents by region

```python
fig2=px.bar(df,x='REGION', y='INCIDENTS',
            title="Incidents by region",
            color="REGION"
           )
fig2.show()
```

# Pie Chart of total incidents by region

```python
fig3 = px.pie(df,values="INCIDENTS", names="REGION",
              title = "Total Incidents by Region",
              width=700,height=400)
fig3.show()
```

# Histogram of estimated damages

```python
fig4 = px.histogram(df,x='ESTIMATED_DAMAGES',nbins=30,
                    title="Distribution of Estimated Damages",
                    color="REGION")
fig4.show()
```

# Scatter plot of Incidents vs Estimated damages

```python
fig5 = px.scatter(df,x='INCIDENTS',y='ESTIMATED_DAMAGES',
                  title="Incidents vs. Estimated Damages",
                  color="YEAR")
fig5.show()
```

# Box plot of estimated damages by region

```python
fig6 = px.box(df,x='REGION',y='ESTIMATED_DAMAGES',
              title="Estimated Damages by Region",
              color="REGION")
fig6.show()
```

# Bubble chart of incidents by region and year

```python
fig7 = px.scatter(df, x='YEAR', y='REGION', size='INCIDENTS',
                  color='REGION', title='Bubble Chart of Incidents by Region and Year', 
                  size_max=60) 
fig7.show()
```

```python
# Heatmap of incidents by year and region 
heatmap_data = df.pivot_table(index='REGION', columns='YEAR', values='INCIDENTS', aggfunc='sum', fill_value=0) 
fig7 = px.imshow(heatmap_data, title='Heatmap of Incidents by Year and Region')
fig7.show() 
```
